## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python

import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("company_sales_data.csv")
plt.figure(figsize=(20,6))
axis1=data["total_profit"]
axis0=data["month_number"]
plt.plot(axis0,axis1,scalex=True)
plt.xticks(range(1,13))
plt.xlabel("Month number")
plt.ylabel("Profit in dollar")
plt.title("Company profit per month")
```

## Exercise 3
**Read all product sales data and show it  using a multiline plot**
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Expected output**

<img src="sales_data.png" alt="sales_data" style="width: 500px;"/>


```python
plt.figure(figsize=(20,5))
data=pd.read_csv("company_sales_data.csv")
sales=data.iloc[:,1:7]
headers=list(sales.columns.values)
plt.plot(sales,marker='o',label=headers)
plt.legend()
plt.xlabel('Month Number')
plt.ylabel('Sale units')
plt.show()
```

## Exercise 4

## Exercise 7
**Read the total profit of each month and show it using the histogram to see the most common profit ranges**

**Expected output**

<img src="Profit_data.png" alt="Profit_data" style="width: 700px;"/>


```python

axis1 = data.loc[:,'total_profit']
plt.figure(figsize=(10,5))
range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(axis1, range,label = 'profit data')
plt.legend(loc ='upper left')
plt.xlabel('profit range in dollar')
plt.ylabel('actual profit')
plt.show()
```

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
circledata  = [data['facecream'].sum(), data['facewash'].sum(), 
data['toothpaste'].sum(), data['bathingsoap'].sum(),data['shampoo'].sum(),data['moisturizer'].sum()]

plt.figure(figsize=(8,5))
plt.pie(circledata, labels=['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer'], autopct='%1.1f%%')
plt.legend(loc= "lower right")
plt.show()
```

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

**Expected output**

<img src="Sales_data_of_a_facewash.png" alt="Sales_data_of_a_facewash" style="width: 700px;"/>


```python

facew  =data.loc[:,'facewash']
baths   = data.loc[:,'bathingsoap']
months = data.loc[:,'month_number']


s,subplt = plt.subplots(2, sharex=True)
subplt[0].plot(months, baths, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
subplt[0].set_title('Sales data of  a Bathingsoap')
subplt[1].plot(months, facew, label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
subplt[1].set_title('Sales data of  a facewash')


plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```
